# Inner join

Use a code recipe or visual recipe to INNER JOIN the manufacturing tables to create a table that has the KPIs (Key Performance Indicators) of each part and the engine serial number (or ESN) they associate with.

```python
import numpy as np
import pandas as pd


#sets the default autosave frequency in seconds
%autosave 60 

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns',None)
#pd.set_option('display.max_rows',None)
pd.set_option('display.width', 1000)
np.random.seed(0)
np.set_printoptions(suppress=True)
```

```python
df = pd.read_csv("avenginedata.csv")
```

```python
df.head(10)
```

```python
df1 = df.sample(frac=0.5,random_state=0)
```

```python
df1
```

# Build a supporting KPI table

```python
suply_chain = pd.read_csv("av_manufacturing_supply_chain_psql.csv")
```

```python
suply_chain
```

```python
suply_chain1 = suply_chain.sample(frac=0.05, random_state=0)
```

```python
suply_chain1
```

```python
av_bom = pd.read_csv("av_bom_manufacturing_psql.csv")
```

```python
av_bom
```

```python
av_bom1 = av_bom.sample(frac=0.1,random_state=0)
```

# Merging table with inner join

```python
kpi_table = pd.merge(left=suply_chain1, right=av_bom1, how='inner',on='pn')
```

```python
kpi_table
```

```python
kpi_table.duplicated().sum()
```

```python
kpi_table1 = kpi_table.sample(frac=0.05, random_state=0)
```

```python
kpi_table1
```

# Join the new KPI table to the consolidated flights table

```python
df1.head(1)
```

```python
kpi_table1.head(1)
```

```python
# joining above two table
df2 = pd.merge(left=df1, right=kpi_table1, how='inner',on='esn')
```

```python
df2 
```

# Join final table with RUL table to get remaining useful life (RUL) for each engine

```python
rul = pd.read_csv("av_esn_rul_psql.csv")
```

```python
rul
```

```python
df1.head(1)
```

```python
final_df = pd.merge(left=df1, right=rul, how='inner',on='esn')
```

```python
final_df
```

```python
final_df.to_csv("final.csv",index=False)
```
